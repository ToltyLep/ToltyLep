```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

G = nx.path_graph(5)

pos = nx.spring_layout(G, seed = 3)

L = nx.laplacian_matrix(G)
L = L.toarray()

Lambda,U = np.linalg.eig(L)

#print(U)

Ordem = np.argsort(Lambda)

#print(Lambda)
#print()
#print(Ordem)

Lambda = Lambda[Ordem]

U = U[:,Ordem]

#print(Lambda)
#print(U)

Coluna = 4
f = U[:,Coluna]

cmap = plt.get_cmap('seismic')

vmin = -1
vmax = 1

plt.figure(dpi=100)
plt.title(r'$\lambda_'+str(Coluna+1)+'$='+str(np.round(Lambda[Coluna],4)))
nx.draw(G,pos, node_color = f, cmap = cmap, vmin = vmin, vmax = vmax)
sm = plt.cm.ScalarMappable(cmap=cmap, norm=plt.Normalize(vmin = vmin, vmax=vmax))
sm._A = []
plt.colorbar(sm)
#plt.savefig('grafo_de_caminho_com_autovetor_'+str(Coluna+1)+'.png')
plt.show()



def cruzamento_por_zero(G, f):
  n_cruzamento_por_zero = 0
  for aresta in G.edges():
    node_0= aresta[0]
    node_1= aresta[1]
    if f[node_0]*f[node_1] < 0:
      n_cruzamento_por_zero += 1
  return(n_cruzamento_por_zero)

#print('nmeros de cruzamentos por zero:',cruzamento_por_zero(G, f))

array_cruzamento_por_zero = np.zeros((len(G),))
for i in range(len(G)):
  array_cruzamento_por_zero[i] = cruzamento_por_zero(G, U[:,i])

#print(array_cruzamento_por_zero)

plt.figure(dpi=100)
plt.title(r'$|Z_G(u_'+str(Coluna+1)+')|$='+str(array_cruzamento_por_zero[Coluna]))
nx.draw(G,pos, node_color = np.sign(f), cmap = cmap, vmin = vmin, vmax = vmax)
sm = plt.cm.ScalarMappable(cmap=cmap, norm=plt.Normalize(vmin = vmin, vmax=vmax))
sm._A = []
plt.colorbar(sm)
plt.savefig('grafo_de_caminho_com_autovetor_'+str(Coluna+1)+'_Zg.png')
plt.show()

plt.figure()
plt.scatter(Lambda,array_cruzamento_por_zero)
plt.savefig('grafo_de_caminho_grafico_de_cruzamentos_por_zero')
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

G = nx.grid_2d_graph(4,4)

pos = dict()
for i in range(4):
  for j in range(4):
    pos[(i,j)] = (i,j)

#pos = nx.spring_layout(G)

L = nx.laplacian_matrix(G)
L = L.toarray()

Lambda,U = np.linalg.eig(L)

#print(U)

Ordem = np.argsort(Lambda)

#print(Lambda)
#print()
#print(Ordem)

Lambda = Lambda[Ordem]

U = U[:,Ordem]

#print(Lambda)
#print(U)

Coluna = 15
f = U[:,Coluna]

cmap = plt.get_cmap('seismic')

vmin = -1
vmax = 1

plt.figure(dpi=100)
plt.title(r'$\lambda_{'+str(Coluna+1)+'}$='+str(np.round(Lambda[Coluna],4)))
nx.draw(G,pos, node_color = f, cmap = cmap, vmin = vmin, vmax = vmax)
sm = plt.cm.ScalarMappable(cmap=cmap, norm=plt.Normalize(vmin = vmin, vmax=vmax))
sm._A = []
plt.colorbar(sm)
#plt.savefig('grafo_de_grade_com_autovetor_'+str(Coluna+1)+'.png')
plt.show()

plt.figure(dpi=100)
#plt.title(r'$\lambda_'+str(Coluna+1)+'$='+str(np.round(Lambda[Coluna],4)))
nx.draw(G,pos, node_color = np.sign(f), cmap = cmap, vmin = vmin, vmax = vmax)
sm = plt.cm.ScalarMappable(cmap=cmap, norm=plt.Normalize(vmin = vmin, vmax=vmax))
sm._A = []
plt.colorbar(sm)
plt.savefig('grafo_de_grade_com_autovetor_'+str(Coluna+1)+'.png')
plt.show()

def cruzamento_por_zero(G, f):
  n_cruzamento_por_zero = 0
  nos = list(G.nodes)
  for aresta in G.edges():
    node_0= aresta[0]
    node_1= aresta[1]

    node_0 = nos.index(node_0)
    node_1 = nos.index(node_1)
    if f[node_0]*f[node_1] < 0:
      n_cruzamento_por_zero += 1
  return(n_cruzamento_por_zero)

#print('nmeros de cruzamentos por zero:',cruzamento_por_zero(G, f))

array_cruzamento_por_zero = np.zeros((len(G),))
for i in range(len(G)):
  array_cruzamento_por_zero[i] = cruzamento_por_zero(G, U[:,i])

print(array_cruzamento_por_zero)
print(np.round(Lambda,3))


plt.figure()
plt.scatter(Lambda,array_cruzamento_por_zero)
plt.scatter(Lambda[Coluna],array_cruzamento_por_zero[Coluna])
plt.savefig('Grafico_grafo_de_grade_'+str(Coluna+1)+'.png')
plt.show()
```
